```python
# Install necessary libraries
!pip install nltk textblob

# Import necessary libraries
import nltk
import pandas as pd
from textblob import TextBlob
from nltk.sentiment import SentimentIntensityAnalyzer
from google.colab import files

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('vader_lexicon')

# Upload the file
uploaded = files.upload()

# Get the name of the uploaded file
file_name = list(uploaded.keys())[0]

# Read the uploaded file (handling both .txt and .csv files)
if file_name.endswith('.txt'):
    with open(file_name, 'r') as file:
        reviews = file.readlines()

elif file_name.endswith('.csv'):
    df = pd.read_csv(file_name)
    reviews = df['review'].tolist()  # Assuming the reviews are in the 'review' column

# Function to perform sentiment analysis using TextBlob
def analyze_sentiment_textblob(review):
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'

# Function to perform sentiment analysis using VADER
sia = SentimentIntensityAnalyzer()
def analyze_sentiment_vader(review):
    sentiment_score = sia.polarity_scores(review)
    compound_score = sentiment_score['compound']
    if compound_score >= 0.05:
        return 'positive'
    elif compound_score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Perform sentiment analysis using TextBlob
textblob_sentiments = [analyze_sentiment_textblob(review) for review in reviews]

# Perform sentiment analysis using VADER
vader_sentiments = [analyze_sentiment_vader(review) for review in reviews]

# Print the results for the first 5 reviews
print("Sentiment Analysis using TextBlob (first 5 reviews):")
for review, sentiment in zip(reviews[:5], textblob_sentiments[:5]):
    print(f"Review: {review} => Sentiment: {sentiment}")

print("\nSentiment Analysis using VADER (first 5 reviews):")
for review, sentiment in zip(reviews[:5], vader_sentiments[:5]):
    print(f"Review: {review} => Sentiment: {sentiment}")

# Save the results to a CSV file
output_df = pd.DataFrame({
    'Review': reviews,
    'TextBlob Sentiment': textblob_sentiments,
    'VADER Sentiment': vader_sentiments
})

output_file = '/content/sentiment_analysis_results.csv'
output_df.to_csv(output_file, index=False)

# Provide a download link for the results
files.download(output_file)
```
